```python
# Using pandas to read data
import pandas
data =  pandas.read_csv("https://modcom.co.ke/data/datasets/schoolcleaned.csv");
data.head()
```

```python
# Create Plots - EDA
# Types of Plots
# Univariate  Plots - Using one column
# Bivariate Plots -   Using two columns
# Multivariate Plot  - More than two column
# Library Matplotlib, seaborn, plotly 
# Import matplotlin and seaborn to be used in our exploratory analysis 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.style.available
```

```python
# pick one style from above
plt.style.use("seaborn")
```

```python
# Create a count plot for Gender
ax = sns.countplot(data['Gender'])
```

```python
# Create a count plot on selfemployed
ax = sns.countplot(data['Smoking'])
```

```python
# Dist plot of Loan Amount
sns.distplot(data['Math'])
```

```python
# Dist plot
sns.distplot(data['Math'], kde  =False)
```

```python
# Property area by Gender
colors = ['#00ff00', '#0000ff','#ffccff']
plt.figure(figsize=(8,6))
ax = sns.countplot(x = 'Gender', palette = 'RdGy_r',
              data=data, hue="Smoking")
```

```python
# Scatter plots - two continous variables

colors = ['#ffff00',"#000000", "#00ff00"]
sns.scatterplot(x = 'Math', y = 'Reading',data = data,
                hue = "State", palette = colors)
```

```python
colors = ['#ffff00',"#000000"]
import numpy as np
sns.barplot(x = 'Gender', y = 'Math',data = data,
                hue = "State", palette = colors, estimator = np.max)
```

```python
plt.style.use("ggplot")
# One continous variable i.e LoanAMount
sns.swarmplot(x = 'Gender', y = 'Math', data=data, 
              hue = "State")
```

```python
plt.style.use("ggplot")
colors = ['#ffff00',"#000000", '#00f000']
# One continous variable i.e LoanAMount, one 2 category variables
sns.swarmplot(x = 'Gender', y = 'Math', data=data, 
              hue = "HowCommute", palette = colors)
```

```python
subset = data[['Height', 'Weight', 'Math', 'Reading', 'Writing', 'Sprint']]
data.corr()
sns.heatmap(subset.corr(), annot=True, cmap= "Greens") # Blues and Reds can also be used
```

```python
# Create pairplot -  all variables by gender
subset = data[['Height', 'Weight', 'Math', 'Reading', 'Writing', 'Sprint', 'Gender']]
sns.pairplot(subset, hue='Gender', size=2.5);
```

```python
sns.stripplot(x='Gender', y='Math', data=data)
# giving title to the plot
plt.title('Graph')
```

```python
# Violin Plot
sns.violinplot(x='Gender', y='Math', data=data,
               hue='State')
```

```python
# Line plot
sns.lineplot(data=data, x="Math", y="Writing", hue = "Gender")
```

```python
sns.boxplot(y = 'State', x = 'Math', data = data)
```

```python
sns.kdeplot( data['Math'], data['Reading'])
```
