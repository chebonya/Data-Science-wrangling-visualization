```python
import pandas  
# Read data, manipulate your data, select columns, basic stats, check empties.
data = pandas.read_csv("https://modcom.co.ke/data/datasets/schoolcleaned.csv", 
                       parse_dates=['bday','enrolldate', 'expgradate'])
data.head(2)
```

```python
# Seaborn, MatplotLib
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.available
```

```python
plt.style.use('ggplot')
# Count, Bar, Box, Line, Swarm, Strip, Point, Heat Maps, KDE, dist,
# Scatter, Violin, pair, 
# Uni variate - Needs only one column
# Bi variate  - Needs two columns
# Multi Variate - Needs more than 2 columns
```

```python
# Count Plot  - Categorical Variables
sns.countplot(data['HowCommute'], palette = 'Oranges')
plt.title("Distribution of HowCommute Status")
plt.xlabel('How they commuted')
plt.ylabel('Number of Students')
```

```python
# Bar
import numpy
# you can use min, max, mean, std, sum, product
# Bi variate, multi variate
ax = sns.barplot(x = 'Gender', y = 'Math', data = data, estimator=numpy.mean,
            palette = 'magma', hue='Athlete')

# justpaste.it/arfz6
for p in ax.patches:
  txt = str(p.get_height().round(1)) + ' '
  txt_x = p.get_x()
  txt_y = p.get_height()
  ax.text(txt_x,txt_y,txt)
```

```python
# Dist
plt.style.use('seaborn')
ax = sns.distplot(data['Sprint'],kde= False)
plt.title("Distribution of Sprint")
plt.xlabel('Sprint in minutes')
plt.ylabel('Freq')
```

```python
# Swarm 
colors = ['#3949AB', '#2E7D32', '#546E7A']
sns.swarmplot(x ='Gender', y ='Math', data = data, palette= 'magma',
              hue ='State')
```

```python
sns.stripplot(x ='Gender', y ='Math', data = data, palette= 'magma',
              hue ='State')
```

```python
sns.violinplot(x ='Gender', y ='Math', data = data, palette= 'magma',
              hue ='State')
```

```python
# Heat Maps
subset = data[['Height', 'Weight', 'Math', 'Reading', 'Writing', 'Sprint']]
subset
sns.heatmap(subset.corr(), annot=True, cmap = 'Blues') # Reds, Greens
```

```python
# Scatter Plot
sns.scatterplot(x ='Math', y ='Writing', data = data, palette= 'magma', hue = 'Athlete')
```

```python
# Box 
# Check for outliers
sns.boxplot(y = 'Reading', data = data, x = 'Athlete')
#data['Reading'].describe()
# IQR = 1.7
# Max = 7.3+1.5*1.7  =9.85
```

```python
# Line Plot
# Scatter Plot
plt.figure(figsize=(20, 5))
sns.lineplot(x ='Math', y ='Reading', data = data, palette= 'magma', hue ='Athlete')
# Try Your own 5 more graphs
```
